#### Importing necessary libraries

```python
import  numpy as np
import pandas as pd
import matplotlib.image as image
import matplotlib.pyplot as plt
import PIL
from PIL import Image
```

#### Loading the images into a sample dictionary

```python
num_list = []
sample_dict = dict()
name = ''
path_start = 'C:\\Users\\Vipra\\OneDrive\\Pictures\\Task_image_identification_numbers\\'
for i in range(0,10):
    for j in range(1,11):
        num_list.append(i)
        name = 'im_no'+'_'+str(i)+'_'+str(j)
        path = path_start + str(i) + '_' + str(j) + '.jpg'
        sample_dict[name] = Image.open(path)    
```

```python
sample_dict
```

```python
print(num_list)
```

#### Converting the images to grayscale using PIL convert and later converting to numpy array

```python
for key,val in sample_dict.items():
    val = val.convert('L')
    val = val.resize((230,129))
    val = np.array(val)
    sample_dict[key] = np.ravel(val)
    
```

```python
sample_dict
```

#### Creating the dataframe from the dictionary and also adding the label column using the num_list 

```python
df_image=pd.DataFrame(sample_dict)
df_image = df_image.T
df_image['label'] = num_list
```

```python
df_image
```

#### Data cleaning- reset_index

```python
df_image.reset_index(drop=True,inplace = True)
```

```python
df_image
```

#### Shuffling the rows of the dataframe and splitting the data into train and test

```python
df_final = df_image.sample(frac = 1).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y = df_final.drop(columns='label'), df_final['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42,stratify=y)
```

#### Visualizing random images with labels to ensure proper labelling is done

```python
trainSample = X.sample(9)
trainSampleIndex = trainSample.index
plt.figure(figsize=(9,9))
for i, val in enumerate(trainSampleIndex):
    plt.subplot(3,3,i+1)
    plt.title(y.iloc[val])
    plt.imshow(X.iloc[val,:].values.reshape(129,230))
```

### Building the ML model- Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
imSVClassifier = SVC()
```

```python
imSVClassifier.fit(X_train,y_train)  ## Fitting x-train and y-train to the model
```

```python
yTrainpred, yTestpred = imSVClassifier.predict(X_train), imSVClassifier.predict(X_test) # predicting the test results
```

#### Generating the confusion_matrix and visualizing the same to understand how many times the classes have been correctly predicted.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
yTestCM = confusion_matrix(y_test, yTestpred, labels = imSVClassifier.classes_)
```

```python
yTestCM
```

```python
cm = ConfusionMatrixDisplay(yTestCM, display_labels=imSVClassifier.classes_)
cm.plot()
```

#### Building a Decision Tree Classifier to compare with the SVC

```python
from sklearn.tree import DecisionTreeClassifier
imDTClassifier = DecisionTreeClassifier()
imDTClassifier.fit(X_train,y_train)
```

```python
yTrainpred, yTestpred = imDTClassifier.predict(X_train), imDTClassifier.predict(X_test)
```

```python
yTestCM = confusion_matrix(y_test, yTestpred, labels = imDTClassifier.classes_)
```

```python
cm = ConfusionMatrixDisplay(yTestCM, display_labels=imDTClassifier.classes_)
cm.plot()
```

`Insights found`: For our problem of image recognition of handwritten numbers, SVC predicted better than DTC
